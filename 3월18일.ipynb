```python
def add_mul(x, y):
    s = x + y
    m = x * y
    return s, m
```

```python
a = add_mul(10, 20)
print(a, type(a))
```

```python
b, c = add_mul(10, 20)
print(b, c)
```

```python
def testArg(*args):
    print(type(args), len(args))
    for n, item in enumerate(args):
        print(n, item)
    return
```

```python
testArg()
testArg(1)
testArg(1, 2, 3)
testArg((1, 2, 3))
```

```python
strings = ['bob', 'charles', 'alexander', 'teddy', 'jane']
strings.sort()
print(strings)
```

```python
def str_len(s):
    return len(s)

strings = ['bob', 'charles', 'alexander', 'teddy', 'jane']
strings.sort(key=str_len)
print(strings)
```

```python
strings = ['bob', 'charles', 'alexander', 'teddy', 'jane']
strings.sort(key=lambda s:len(s))
print(strings)
```

```python
def even(n):
    return n%2==0

a = [10, 13, 2, 34, 8, 90, 11]
b = list(filter(even, a))
print(b, type(b))
```

```python
def averaging(input):
    sum = 0
    for value in input:
        sum += value
    mean = sum / len(input)
    return mean
```

```python
avg_a = averaging(a)
print(avg_a)
```

num%2 != 0, num%3 !=0, num%4 !=0, ..., num%(num-1) != 0 ==> prime number

```python
def isPrime(num):
    for i in range(2, num):
        if num%i == 0:
            return False
    return True
```

```python
isPrime(53)
```

```python
a = []
a.append(1)
print(a)
a.append(2)
print(a)
```

```python
def findPrime(num):
    answer = []
    for i in range(2, num+1):
        if isPrime(i)==True:
            answer.append(i)
    return answer
```

```python
c = findPrime(40)
print(c)
```

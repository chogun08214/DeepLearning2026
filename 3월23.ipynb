## Class, Object

```python
class Person: 
    def __init__(self, name_in, age_in = 10):
        self.name = name_in
        self.age = age_in
        self.hobby = 'singing'
    def PrintAll(self):
        print(self.name, self.age, self.hobby)
```

```python
p1 = Person('bob', 30)
p2 = Person('kang', 40)
p3 = Person('kim')
```

```python
print(p1.name, p1.age, p1.hobby)
print(p2.name, p2.age, p2.hobby)
print(p3.name, p3.age, p3.hobby)
p1.PrintAll()
p2.PrintAll()
p3.PrintAll()
```

### inheritance

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def Eat(self, food):
        print(f'{self.name}은 {food}를 먹는다')
    def Sleep(self, hours):
        print(f'{self.name}은 {hours}시간을 잔다')
    def Work(self, hours):
        print(f'{self.name}은 {hours}시간을 일한다')
```

```python
bob = Person('Bob', 30)
bob.Eat('BBQ')
bob.Sleep(8)
bob.Work(10)
```

```python
class Student(Person):
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def Work(self, hours):
        super().Work(hours)
        print(f'{self.name}은 {hours}시간을 공부한다')
class Employee(Person):
    def __init__(self, name, age):
        self.name = name
        self.age = age
```

```python
bob = Student('Bob', 30)
bob.Eat('BBQ')
bob.Sleep(8)
bob.Work(10)
```

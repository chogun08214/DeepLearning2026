```python
import numpy as np
```

```python
X = np.array([[-1, -1], [-1, 1], [1, -1], [1, 1]])
z = np.array([[-1], [1], [1], [1]])
print(X, X.shape)
print(z, z.shape)
```

```python
def sigmoid(y):
    return 2.0/(1+np.exp(-y)) - 1.0
```

```python
sigmoid(np.array([0, 1, 2]))
```

```python
def sigmoid_derivative(y):
    z = sigmoid(y)
    return 0.5 * (1-z) * (1+z)
```

```python
np.random.seed(1000)
w = np.random.randn(2, 1)
b = np.zeros((1, 1))

print(w, w.shape)
print(b, b.shape)
```

```python
layer0 = X
y = np.dot(X, w) + b
layer1 = sigmoid(y)
print(y, y.shape)
print(np.dot(X, w))
print(layer1)
```

```python
def forward_compute(X, w, b):
    layer0 = X
    y = np.dot(X, w) + b
    layer1 = sigmoid(y)
    return layer0, layer1
```

```python
np.dot(X, w), X.shape, w.shape
```

```python
learning_rate = 0.1
layer0, layer1 = forward_compute(X, w, b)

decision = layer1.copy()
decision[decision>=0] = 1
decision[decision<0] = -1
error = np.sum(np.abs(decision-z))
loss = np.sum(np.abs(layer1 - z))
print(error, loss, layer1.T)
#if error == 0: break

layer1_error = layer1 - z
layer1_delta = layer1_error * sigmoid_derivative(layer1)

print(layer1_error)
print(layer1_delta)
```

```python
print(X.T)
print(X.T.shape, layer1_delta.shape)
print(len(X))
```

```python
dw = np.dot(X.T, layer1_delta) / len(X)
db = np.sum(layer1_delta) / len(X)

w -= learning_rate * dw
b -= learning_rate * db
```

```python
learning_rate = 0.1
for epoch in range(1000):
    layer0, layer1 = forward_compute(X, w, b)

    decision = layer1.copy()
    decision[decision>=0] = 1
    decision[decision<0] = -1
    error = np.sum(np.abs(decision-z))
    loss = np.sum(np.abs(layer1 - z))
    print(epoch, int(error), loss, layer1.T)
    #if error == 0: break

    layer1_error = layer1 - z
    layer1_delta = layer1_error
    #layer1_delta = layer1_error * sigmoid_derivative(layer1)

    dw = np.dot(X.T, layer1_delta) / len(X)
    db = np.sum(layer1_delta) / len(X)

    w -= learning_rate * dw
    b -= learning_rate * db
```

```python
a, b, c = 10, 20, 30
if (a==10 and b==20) or c==100:
    print('1번째 조건이 참이다')
else:
    print('1번째 조건이 거짓이다')

if a!=10:
    print('2번째 조건이 참이다')
else:
    print('2번째 조건이 거짓이다')

if not a==20:
    print('3번째 조건이 참이다')
```

```python
a = 10
if a%4 == 0:
    print('{}는 나머지가 0이다'.format(a))
elif a%4 == 1:
    print('{}는 나머지가 1이다'.format(a))
elif a%4 == 2:
    print('{}는 나머지가 2이다'.format(a))
else:
    print('{}는 나머지가 3이다'.format(a))
```

```python
a = [10, 20, 30, 40, 50, 60, 70]
for value in a:
    if value > 30:
        break
    print(value)
```

```python
a = [10, 20, 30, 40, 50, 60, 70]
for value in a:
    if value > 30 and value < 70:
        continue
    print(value)
```

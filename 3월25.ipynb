```python
class Point2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __add__(self, pt):
        new_x = self.x + pt.x
        new_y = self.y + pt.y
        return Point2D(new_x, new_y)
    def __str__(self):
        return '({}, {})'.format(self.x, self.y)    
```

```python
p1 = Point2D(3, 5)
p2 = Point2D(-1, 1)
```

```python
p3 = p1 + p2
```

```python
print(p1, p2, p3)
```
